# Sales Data Analysis using Python

This project analyzes sales data to understand revenue trends and product performance using Python.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales_df = pd.DataFrame({
    "order_id": [1, 2, 3, 4, 5],
    "order_date": [
        "2024-01-05",
        "2024-01-10",
        "2024-02-15",
        "2024-02-20",
        "2024-03-05"
    ],
    "product_name": [
        "Laptop",
        "Mobile Phone",
        "Headphones",
        "Mobile Phone",
        "Laptop"
    ],
    "price": [60000, 25000, 3000, 25000, 60000],
    "quantity": [1, 2, 3, 1, 1]
})

sales_df["revenue"] = sales_df["price"] * sales_df["quantity"]
sales_df
```

```python
sales_df["order_date"] = pd.to_datetime(sales_df["order_date"])
sales_df["month"] = sales_df["order_date"].dt.month
sales_df
```

```python
monthly_revenue = (
    sales_df
    .groupby("month")["revenue"]
    .sum()
    .reset_index()
)

monthly_revenue
```

```python
plt.figure()
plt.plot(monthly_revenue["month"], monthly_revenue["revenue"])
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Monthly Revenue Trend")
plt.show()
```

```python
product_revenue = (
    sales_df
    .groupby("product_name")["revenue"]
    .sum()
    .reset_index()
)

product_revenue
```

```python
plt.figure()
plt.bar(product_revenue["product_name"], product_revenue["revenue"])
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.title("Revenue by Product")
plt.show()
```

```python
total_revenue = sales_df["revenue"].sum()
total_revenue
```

## Key Insights

- Laptops generated the highest revenue.
- January was the best-performing month.
- Total revenue generated was 204,000.
